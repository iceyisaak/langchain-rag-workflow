# VectorStore: ChromaDB

### Load: TextLoader

```python
from langchain_community.document_loaders import TextLoader

text_loader = TextLoader('speech.txt')
text_document = text_loader.load()
```

```python
text_document
```

### Split: RecurCharSplitter

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

recurchar_splitter = RecursiveCharacterTextSplitter(chunk_size=500,chunk_overlap=0)
splitted_docs = recurchar_splitter.split_documents(text_document)
```

```python
splitted_docs
```

### Embed: OllamaEmbeddings

```python
from langchain_community.embeddings import OllamaEmbeddings

ollama_embeddings = OllamaEmbeddings()
```

### Store: ChromaDB

```python
from langchain_chroma import Chroma
```

```python
chroma_db = Chroma.from_documents(splitted_docs, ollama_embeddings)
```

```python
chroma_db
```

### Query: Similarity_Search()

```python
query='How does the speaker describe the outcome of the war?'

query_results = chroma_db.similarity_search(query)
```

```python
query_results
```

```python
query_results[0].page_content
```

### Query: Search Results by Similarity Score

```python
query_results_by_similarity_score = chroma_db.similarity_search_with_score(query)

query_results_by_similarity_score
```

### Query: Retriever

```python
retriever = chroma_db.as_retriever()

retriever_results = retriever.invoke(query)
```

```python
retriever_results
```

```python
retriever_results[0].page_content
```

### Save vector to disk

```python
chroma_db = Chroma.from_documents(documents=splitted_docs, embedding=ollama_embeddings,persist_directory='./chroma_db')
```

### Load vector from disk

```python
new_db = Chroma(persist_directory='./chroma_db',embedding_function=ollama_embeddings)
```

```python
new_docs = new_db.similarity_search(query)

new_docs
```
