# VectorStores - FAISS (Facebook AI Similarity Search)

### Load: TextLoader

```python
from langchain_community.document_loaders import TextLoader

text_loader = TextLoader('speech.txt')
text_document = text_loader.load()
```

```python
text_document
```

### Split: CharTextSplitter

```python
from langchain_text_splitters import CharacterTextSplitter

char_splitter = CharacterTextSplitter(chunk_size=1000,chunk_overlap=30)
docs = char_splitter.split_documents(text_document)
```

```python
docs
```

### Embed: OllamaEmbeddings

```python
from langchain_community.embeddings import OllamaEmbeddings

ollama_embeddings = OllamaEmbeddings()
```

### Store: FAISS

```python
# import faiss
from langchain_community.vectorstores import FAISS
```

```python
vector_db = FAISS.from_documents(docs,ollama_embeddings)
```

### Query: .similarity_search()

```python
query='How does the speaker describe the outcome of the war?'

query_result = vector_db.similarity_search(query)
```

```python
query_result
```

```python
query_result[0].page_content
```

### Query from Retriever: .as_retriever()

```python
retriever = vector_db.as_retriever()

retriever_result = retriever.invoke(query)
```

```python
retriever_result
```

```python
retriever_result[0].page_content
```

### Query: FAISS with Search Score (Score = L2 Distance; Lower Score is better)

```python
docs_with_score = vector_db.similarity_search_with_score(query)

docs_with_score
```

```python
query_embeded = ollama_embeddings.embed_query(query)

query_embeded
```

```python
query_results_by_score = vector_db.similarity_search_by_vector(query_embeded)

query_results_by_score
```

---

### Saving FAISS Index

```python
vector_db.save_local('faiss_index')
```

### Loading FAISS Index

```python
new_db = FAISS.load_local(
    'faiss_index',
    ollama_embeddings,

    # Safety Lock
    allow_dangerous_deserialization=True
)
```

```python
new_docs = new_db.similarity_search(query)

new_docs
```
