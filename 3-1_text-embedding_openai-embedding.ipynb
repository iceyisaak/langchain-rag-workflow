### Read Text Document

```python
from langchain_community.document_loaders import TextLoader
```

```python
text_loader = TextLoader('speech.txt')
```

```python
text_document = text_loader.load()
```

```python
text_document
```

### Recursive Character Splitting

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

```python
recurchar_splitter = RecursiveCharacterTextSplitter(chunk_size=500,chunk_overlap=50)
```

```python
final_docs = recurchar_splitter.split_documents(text_document)
```

```python
final_docs
```

### Vector Embedding & Vector Store

```python
from langchain_community.vectorstores import Chroma
```

```python
vectorstore_db = Chroma.from_documents(final_docs,openai_embeddings_1024)
```

```python
vectorstore_db
```

### Similarity Search

```python
query='It will be all the easier for us to conduct ourselves as belligerents'
```

```python
similarity_search_results = vectorstore_db.similarity_search(query)
```

```python
similarity_search_results
```
