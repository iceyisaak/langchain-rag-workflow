## HuggingfaceEmbeddings

Model: https://huggingface.co/sentence-transformers/all-MiniLM-L6-v2

```python
from langchain_huggingface import HuggingFaceEmbeddings
```

```python
huggingface_embeddings = HuggingFaceEmbeddings(model_name='all-MiniLM-L6-v2')
```

```python
texts='Testing document for huggingface embeddings'
```

```python
query_results = huggingface_embeddings.embed_query(texts)
```

```python
query_results
```

```python
len(query_results)
```

```python
document_results = huggingface_embeddings.embed_documents([texts, 'This is NOT a test document.'])
```
