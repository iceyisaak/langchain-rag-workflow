# Ollama Embeddings

```python
from langchain_community.embeddings import OllamaEmbeddings
```

```python
ollama_embeddings = (
    OllamaEmbeddings(
        model='gemma:2b'
))
```

```python
ollama_embeddings
```

---

```python
texts=[
    'OpenAI is NOT an opensourced tool',
    'Ollama is an opensourced tool'
]
```

```python
embedded_texts = ollama_embeddings.embed_documents(texts)
```

```python
embedded_texts
```

```python
len(embedded_texts[0])
```

```python
texts='Which llm is opensourced?'
```

```python
embedded_texts = ollama_embeddings.embed_query(texts)
```
